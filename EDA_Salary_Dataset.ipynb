## 1. Import Required Libraries

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load Dataset

```python

df = pd.read_csv("Salary_Data.csv")
df.head()
```

## 3. Dataset Shape and Information

```python

print("Shape of dataset:", df.shape)
df.info()
```

## 4. Column-wise Data Types

```python

df.dtypes
```

## 5. Missing Value Analysis

```python

missing = df.isnull().sum()
missing
```

```python
# frequency

df[['Education Level']].value_counts()
```

## 6. Statistical Summary (Numerical Features)

```python

df.describe()
```

## 7. Univariate Analysis

### 7.1 Age Distribution

```python

plt.figure()
plt.hist(df["Age"].dropna())
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### 7.2 Salary Distribution

```python

plt.figure()
plt.hist(df["Salary"].dropna())
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

### 7.3 Gender Distribution

```python

plt.figure()
df["Gender"].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

## 8. Outlier Detection using Boxplot

```python

plt.figure()
plt.boxplot(df["Salary"].dropna())
plt.title("Salary Boxplot")
plt.ylabel("Salary")
plt.show()
```

## 9. Bivariate Analysis

### 9.1 Salary vs Years of Experience

```python

plt.figure()
plt.scatter(df["Years of Experience"], df["Salary"])
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.show()
```

### 9.2 Salary vs Education Level

```python

plt.figure()
df.boxplot(column="Salary", by="Education Level", rot=45)
plt.title("Salary by Education Level")
plt.suptitle("")
plt.show()
```

## 10. Correlation Analysis

```python

corr = df[["Age", "Years of Experience", "Salary"]].corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix")
plt.show()

corr
```
