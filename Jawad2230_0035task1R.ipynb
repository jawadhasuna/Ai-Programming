```r

# Install ggplot2 if not already
install.packages("ggplot2")
library(ggplot2)

# Histogram of Sepal.Length
ggplot(iris, aes(x = Sepal.Length)) +
  geom_histogram(binwidth = 0.5, fill="skyblue", color="black") +
  ggtitle("Distribution of Sepal Length")

# Boxplot of Sepal.Length by Species
ggplot(iris, aes(x = Species, y = Sepal.Length, fill = Species)) +
  geom_boxplot() +
  ggtitle("Sepal Length by Species")

# Pairwise scatterplots for numeric columns
pairs(iris[1:4], main="Pairwise Scatterplots of Iris Features", pch=21, bg=c("red","green","blue")[unclass(iris$Species)])
```
