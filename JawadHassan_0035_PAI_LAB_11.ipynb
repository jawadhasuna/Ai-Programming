```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#task1
grades = ['A', 'B', 'C', 'B', 'A', 'C', 'C', 'B', 'A', 'A', 'B']

# Create DataFrame
df = pd.DataFrame({'grades': grades})

# Calculate frequency
frequency = df['grades'].value_counts()
print(frequency)

# Bar Chart (categorical data)
plt.figure(figsize=(6,4))
plt.bar(frequency.index, frequency.values, color='skyblue', edgecolor='black')
plt.title("Bar Chart of Grades")
plt.xlabel("Grades")
plt.ylabel("Number of Students")
plt.show()

# Histogram (using pandas plot)
plt.figure(figsize=(6,4))
df['grades'].hist(bins=[0,1,2,3,4], color='lightgreen', edgecolor='black')
plt.title("Histogram of Grades")
plt.xlabel("Grades")
plt.ylabel("Number of Students")
plt.xticks([0,1,2,3,4], ['A','B','C','D','E'])  # Map bins to grade labels
plt.show()
```

```python
#task2
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import string
# Given text
text = """Airports are always a bustling entity with many emotions all over the place. Some are happy to
welcome their loved ones in their arms, some are sad to see them go, while others are anxiously waiting
for their flights to reach. Rolling suitcases, hurrying footsteps, laughing children and the electronic ding
of the machines, all accumulated in one place."""

# Convert to lowercase and keep only alphabets
cleaned_text = ''.join([char.lower() for char in text if char.isalpha()])
print(cleaned_text)
# Count frequency of each letter
letter_freq = Counter(cleaned_text)
print(letter_freq)
# Convert to DataFrame for easier plotting
df = pd.DataFrame.from_dict(letter_freq, orient='index', columns=['Frequency']).sort_index()
print(df)
# Bar Chart (categorical data)
plt.figure(figsize=(12,6))
plt.bar(df.index, df['Frequency'], color='skyblue', edgecolor='black')
plt.title("Bar Chart of letters")
plt.xlabel("letters")
plt.ylabel("freq")
plt.show()
```

```python
#task3
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt

# Given data
data = [23, 45, 56, 34, 78, 90, 12, 67, 43, 55, 88, 32, 67, 54, 78, 89, 22, 35, 66, 43]
# Mean
mean = np.mean(data)

# Median
median = np.median(data)

# Mode (may be more than one)
mode = stats.multimode(data)  # Returns list of modes

# Variance
variance = np.var(data, ddof=0)  # Population variance
sample_variance = np.var(data, ddof=1)  # Sample variance

# Standard Deviation
std_dev = np.std(data, ddof=0)  # Population std
sample_std = np.std(data, ddof=1)  # Sample std

# Range
data_range = max(data) - min(data)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance (Population):", variance)
print("Variance (Sample):", sample_variance)
print("Standard Deviation (Population):", std_dev)
print("Standard Deviation (Sample):", sample_std)
print("Range:", data_range)
# Quartiles
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)

# Interquartile Range (IQR)
iqr = q3 - q1

print("Q1 (25th percentile):", q1)
print("Q2 (Median / 50th percentile):", q2)
print("Q3 (75th percentile):", q3)
print("IQR:", iqr)
plt.figure(figsize=(8,6))
plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Boxplot of the Data")
plt.ylabel("Values")
plt.show()
```

```python
#task4

# Load the dataset
df = pd.read_csv('/content/laptops - laptops.csv')
print(df.isnull().any())
# Display first 5 rows (header)
print("Header:")
print(df.head())

# Information about dataset
print("\nInfo:")
print(df.info())

# Statistical summary
print("\nDescription:")
print(df.describe())

import matplotlib.pyplot as plt

# Assuming there is a column 'Brand' in the dataset
brand_counts = df['Brand'].value_counts()
print("\nNumber of brands:", df['Brand'].nunique())
print("\nBrand counts:")
print(brand_counts)

# Bar chart for brands
plt.figure(figsize=(12,6))
plt.bar(brand_counts.index, brand_counts.values, color='skyblue', edgecolor='black')
plt.title("Number of Laptops per Brand")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Assuming the price column is 'Price'
prices = df['Final Price']

# Calculations
mean_price = np.mean(prices)
median_price = np.median(prices)
mode_price = stats.multimode(prices)  # Returns list of modes
variance_price = np.var(prices, ddof=0)  # Population variance
sample_variance = np.var(prices, ddof=1)  # Sample variance

print("Mean Price:", mean_price)
print("Median Price:", median_price)
print("Mode Price:", mode_price)
print("Variance (Population):", variance_price)
print("Variance (Sample):", sample_variance)

# Boxplot
plt.figure(figsize=(8,6))
plt.boxplot(prices, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title("Boxplot of Laptop Prices")
plt.ylabel("Price")
plt.show()
```
