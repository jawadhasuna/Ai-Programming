```python
#t6
import re
text = """
User: John Doe
Email: john.doe@gmail.com
Backup Email: johnd123@company.org
IP: 192.168.0.1
Invalid IP: 999.888.77.666
Date Joined: 12-05-2023
Website: https://www.example.com/profile
Contact: +92-300-1234567 or (051) 6543210
"""
pemial=r'\w+.\w+@\w+.\w+'
emails=re.findall(pemial,text)
pip=r'\d{3}\.\d{3}\.\d{1}\.\d{1}'
ips=re.findall(pip,text)
pd=r'\d{2}-\d{2}-\d{4}'
date=re.findall(pd,text)
pp=r'(\+\d{2}-\d{3}-\d+|\(\d{3}\)\s*\d{7})'
phone=re.findall(pp,text)
pu=r'\w+\:\//\w+\.\w+\.\w+\/\w+'
urls=re.findall(pu,text)
dic={
    "emails": emails,
    "ips": ips,
    "dates": date,
    "phones": phone,
    "urls": urls

}
print(dic)
```
