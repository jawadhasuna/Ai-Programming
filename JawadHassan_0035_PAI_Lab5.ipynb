```python
def maxl(l):

  if len(l)==1:
    return l[0]
  sub=maxl(l[1:])
  return l[0] if l[0]>sub else sub


maxl([1,2,3,4,5])
```

```python
def bins(n,target,low,high):
  if low>high:
     return -1
  mid = (low + high) // 2
  if n[mid]==target:
    return mid
  elif n[mid]>target:
    return bins(n,target,low,mid-1)
  else:
    return bins(n,target,mid+1,high)
num=[2,4,6,8,10,12,14,16,18]
bins(num,14,0,len(num)-1)
```

```python
#T3
def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)  # recursive call if it's a list
        else:
            total += item              # add directly if it's a number
    return total

# Example:
nums = [1, [2, [3, 4]], 5]
print("Sum:", sum_nested(nums))
```

```python
#T4
def dec_to_bin(n):
    if n == 0:
        return ""
    else:
        return dec_to_bin(n // 2) + str(n % 2)

# Example:
num = 13
binary = dec_to_bin(num)
print("Binary of", num, "is:", binary)
```

```python
#T5
def flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))   # recursive call for sublist
        else:
            result.append(item)
    return result

# Example:
nums = [1, [2, [3, 4]], 5]
print("Flattened list:", flatten(nums))
```

```python
#T8
def permute(s, answer=""):
    if len(s) == 0:                     # base case
        print(answer)
        return

    for i in range(len(s)):             # pick each character one by one
        ch = s[i]
        left = s[:i] + s[i+1:]          # remaining string after removing ch
        permute(left, answer + ch)      # recursive call


# Example:
permute("abc")
```
