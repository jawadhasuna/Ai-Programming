## Question 1

```python
def palin(st):
  stt = ""
  for ch in st:
        if ch.isalpha():     # allow only letters
            stt += ch.lower()
  if len(stt)<=1:
    return True
  if stt[0]!=stt[-1]:
    print('not palindrome')
    return False
  return palin(stt[1:-1])

print(palin('mada 55m'))
```

## Question 2

```python
lst = [1, [2, [3, 4], 5], 6]
def sum_nested_list(a):
    total = 0
    for x in a:
        if isinstance(x, list):      # if element is a list â†’ recurse
            total += sum_nested_list(x)
        else:
            total += x               # normal number
    return total

sum_nested_list(lst)
```

```python
def process_numbers(nums):
  a=set()
  for n in nums:
    if n not in a:
      a.add(n)
  print(a)
  a=list(a)
  a.sort()
  mx=max(a)
  mn=min(a)
  s=sum(a)
  avg =sum(a)/len(a)
  aa=(mx,mn,s,avg)
  return aa
def main():
  nums=[]
  for i in range(15):
   nums.append(int(input('enter 15 numbers:')))
  tup= process_numbers(nums)
  print(tup)
main()
```

```python
import string
with open("data.txt",'w') as f:
    f.write("""Python is a popular programming language! PYTHON is easy to learn, and powerful.
Python can be used for web development; data analysis, Artificial Intelligence, and more...
Learning Python helps you automate tasks, solve problems efficiently.
Python, python; PYTHON?""")
with open("data.txt",'r') as f:
  text=f.read()
text=text.lower()
text = text.translate(str.maketrans("", "", string.punctuation))
words = text.split()
wordc={}
for word in words:
  if word in wordc:
    wordc[word]+=1
  else:
    wordc[word]=1
print(wordc)
top_5 = sorted(wordc.items(), key=lambda x: x[1], reverse=True)[:5]
# Step 6: Print top 5 words with counts
print("Top 5 most frequent words:")
for word, count in top_5:
    print(word, count)
```

```python
dic={'Ali': 85,  'Sara': 92, 'Bilal': 45}
newdic={}
for key,value in dic.items():
  if value>=90:
    newdic[key]='pass w distinction'
  elif value>=80:
    newdic[key]='pass'
  else:
    newdic[key]='fail'
print(newdic)
# Count students in each category
category_count = {}
for status in newdic.values():
    if status in category_count:
        category_count[status] += 1
    else:
        category_count[status] = 1

print("Student counts in each category:")
for category, count in category_count.items():
    print(f"{category}: {count}")
```
