```python
#q2
class appoinments:
  def __init__(self):
    self.patient=''
    self.doctor=''
    self.time=''
    self.app=[]
  def add_app(self,p,d,t):
     if t in [a[2] for a in self.app]:
      print(f"Time {t} already booked!")
      return
     self.patient=p
     self.doctor=d
     self.time=t
     self.app.append([self.patient,self.doctor,self.time])
  def cancel_app(self,patient):
    try:
      if not any(a[0] == patient for a in self.app):
       raise Exception("Patient does not exist")
    except Exception as e:
      print(e)
    else:
      for a in self.app:
       if a[0] == patient:
        self.app.remove(a)
        break
  def show(self):
      for i in self.app:
        print(i)

p1=appoinments()
p1.add_app('jawd','d1','10')
p1.add_app('ali','d2','11')
p1.show()

p1.cancel_app('Ali')
p1.show()
```

```python
#q3
class airline:
  def __init__(self):
    self.dic={"Economy": 50, "Business": 10, "First Class": 5}
  def book(self,cat):
    for key,val in self.dic.items():
      if cat==key:
        self.dic[key]-=1

  def cancel(self,cat):
    for key,val in self.dic.items():
      if cat==key:
        self.dic[key]+=1

  def show(self):
    for key,val in self.dic.items():
      print(key,":",val)
a = airline()
a.book("Economy")
a.book("First Class")
a.show()          # show all
a.cancel("Economy")
a.show()
```

```python
#q4


# Step 2: Import os for file handling
import os

# Step 3: Recursive search function
def search_file(directory, keyword):
    try:
        for item in os.listdir(directory):
            full_path = os.path.join(directory, item)

            # If it's a file, check for keyword (case-insensitive)
            if os.path.isfile(full_path):
                if keyword.lower() in item.lower():
                    print(full_path)

            # If it's a directory, recurse into it
            elif os.path.isdir(full_path):
                search_file(full_path, keyword)

    except FileNotFoundError:
        print(f"Error: Directory '{directory}' not found.")
    except PermissionError:
        print(f"Error: Permission denied for directory '{directory}'.")

# Step 4: Use the function on a folder in your Drive
# Replace 'YourFolder' with the folder name you want to search
search_file("/content", "papers.txt")

```

```python
import csv
import re

# Sample employee data
employees = [
    {"Name": "Sara Khan", "Email": "sara.khan@company.com", "Phone": "03451234567"},
    {"Name": "John Doe", "Email": "john@doe", "Phone": "12345"}
]

# Write employees.csv
try:
    with open('employees.csv', 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=['Name', 'Email', 'Phone'])
        writer.writeheader()
        writer.writerows(employees)
except IOError:
    print('File write error')

print("employees.csv created successfully!\n")

# Regex patterns
email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
phone_pattern = r'^03\d{9}$'  # Pakistani mobile numbers

# Counters
total = 0
valid_count = 0
invalid_count = 0

# Store invalid records
invalid_records = []

# Read CSV and validate
try:
    with open('employees.csv', 'r', newline='') as f:
        reader = csv.DictReader(f)
        for row in reader:
            total += 1
            name = row['Name'].strip()
            email = row['Email'].strip()
            phone = row['Phone'].strip()

            # Validate email and phone
            email_valid = bool(re.match(email_pattern, email))
            phone_valid = bool(re.match(phone_pattern, phone))

            if email_valid and phone_valid:
                valid_count += 1
                print(f"{name}: Email=Valid, Phone=Valid")
            else:
                invalid_count += 1
                print(f"{name}: Email={'Valid' if email_valid else 'Invalid'}, Phone={'Valid' if phone_valid else 'Invalid'}")
                invalid_records.append({'Name': name, 'Email': email, 'Phone': phone})

except FileNotFoundError:
    print("employees.csv not found!")
except Exception as e:
    print("Error:", e)

# Write invalid records to a file
if invalid_records:
    with open('invalid_records.txt', 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=['Name', 'Email', 'Phone'])
        writer.writeheader()
        writer.writerows(invalid_records)
    print("\nInvalid records written to invalid_records.txt")

# Print summary
print("\nSummary:")
print(f"Total Records: {total}")
print(f"Valid: {valid_count}")
print(f"Invalid: {invalid_count}")
```

```python
#q8
class lib:
  def __init__(self):
    self.name=''
    self.books=[]
  def add(self,title):
    self.books.append(title)
  def remove(self,title):
    if title in self.books:
      self.books.remove(title)
  def save(self):
    with open('lib.txt','w') as f:
      for b in self.books:
        f.write(b+'\n')
  def __del__(self):
    with open('lib.txt','w') as f:
      for b in self.books:
        f.write(b+'\n')

l = lib()

l.add("Harry Potter")
l.add("LOTR")
l.add("Percy Jackson")

l.remove("LOTR")   # removes

l.save()
l.add("LOTR")
del l

```
