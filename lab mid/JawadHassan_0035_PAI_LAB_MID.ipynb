```python
class Bus:
    def __init__(self, bid,btype):
        self.bid = bid
        self.btype=btype
        self.seats = []  # list to store seat bookings

    def seatbook(self):
    
        try:
            sno = int(input('Enter seat number: '))
        except ValueError:
            print("Invalid seat number!")
            return

        stype = input('Enter seat type (business~high fare 2000/economy~low fare500): ').lower()
        if stype == 'business':
            fare = 2000
            acc=fare*0.5
        elif stype == 'economy':
            fare = 500
            acc=fare*0.5
            
        else:
            print("Invalid seat type!")
            return

        try:
            bstatus = input("Enter booking status (vacant/booked): ").lower()
            if bstatus not in ['vacant', 'booked']:
                raise ValueError("Status must be 'vacant' or 'booked'.")
            else:
                acc=acc*2
        except ValueError as e:
            print("Error:", e)
            return

        seat_info = {
            'sno': sno,
            'btype': stype,
            'fare': fare,
            'pay':acc,
            'status': bstatus
        }
        self.seats.append(seat_info)
        print("Seat booked successfully:", seat_info)
    
    def cancel_booking(self):
        try:
            sno = int(input("Enter seat number to cancel: "))
        except ValueError:
            print("Invalid seat number!")
            return

        for seat in self.seats:
            if seat['sno'] == sno and seat['status'] == 'booked':
                seat['status'] = 'vacant'
                seat['pay']=seat['pay']*0.25
                print(f"Booking cancelled for seat {sno}")
                return
        print(f"No booked seat found with seat number {sno}")

    def handle_payment(self):
        total = sum(seat['fare'] for seat in self.seats if seat['status'] == 'booked')
        print(f"Total payment made for Bus {self.bid}: {total}")
        for seat in self.seats :
            if seat['status'] == 'vacant':
                print('refund:',seat['pay'])


class Transport:
    def __init__(self):
        self.buses = []

    def add_bus(self, bus_obj):
        self.buses.append(bus_obj)

    def display_buses(self):
        if not self.buses:
            print("No buses available.")
            return
        for bus in self.buses:
            print(f"\nBus ID: {bus.bid}")
            if not bus.seats:
                print("No seats booked yet.")
            else:
                for seat in bus.seats:
                    print(seat)


# Example Usage
def main():
    transport_system = Transport()

    # Adding buses
    bid=int(input('enter bus id:'))
    try:
     btype=input('enter bus type: sleeper or standard').lower()
     if btype not in ['sleeper', 'standard']:
                raise ValueError("type must be 'sleeper' or 'standard'.")
    except ValueError as e:
        print("Error:", e)
        return
    
    bus1 = Bus(bid,btype)
    bus1.seatbook()
    bus1.handle_payment()
    bid=int(input('enter bus id:'))
    try:
     btype=input('enter bus type: sleeper or standard').lower()
     if btype not in ['sleeper', 'standard']:
                raise ValueError("type must be 'sleeper' or 'standard'.")
    except ValueError as e:
        print("Error:", e)
        return
    bus2 = Bus(bid,btype)
    bus2.seatbook()
    bus2.handle_payment()
    transport_system.add_bus(bus1)
    transport_system.add_bus(bus2)
    transport_system.display_buses()
    
main()
```
